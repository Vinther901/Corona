```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.manifold import TSNE
import umap
```

```python
confirmed_dat = pd.read_csv('data/time_series_covid19_confirmed_US.csv')
deaths_dat = pd.read_csv('data/time_series_covid19_deaths_US.csv')
```

```python
dateCols = confirmed_dat.columns[11:]
```

```python
confirmed_dated = confirmed_dat[['Province_State']+list(dateCols)]
confirmed_dated.head()
```

```python
maximums = np.max(confirmed_dated,axis=1)
for i in range(len(confirmed_dated)):
    try:
        confirmed_dated.iloc[i,1:] /= maximums[i]
    except ZeroDivisionError:
        pass
```

```python
clean_normalized = confirmed_dated.dropna().reset_index(drop=True)
```

```python
fft = np.fft.rfft(clean_normalized.iloc[:,1:])
```

```python
help(TSNE)
```

```python
help(umap.UMAP)
```

```python
emb_tsne = TSNE(verbose=2,n_iter=25000).fit_transform(abs(fft))
```

```python
emb_umap = umap.UMAP(verbose=2,n_epochs=6000).fit_transform(abs(fft))
```

```python
plt.figure()
plt.scatter(emb_tsne[:,0],emb_tsne[:,1],s=10,c = np.log(maximums[maximums != 0]),cmap='plasma')

plt.figure()
plt.scatter(emb_umap[:,0],emb_umap[:,1],s=10,c = np.log(maximums[maximums != 0]),cmap='plasma')
```
