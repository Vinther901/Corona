```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.manifold import TSNE
import umap
```

```python
dat = pd.read_csv('Data/time_series_covid19_confirmed_global.csv')
dat.head()
```

```python
dat_dated = dat[['Country/Region'] + list(dat.columns[4:])]
```

```python
maximums = np.max(dat_dated.iloc[:,1:],axis=1)
```

```python
for i in range(len(dat_dated)):
    if maximums[i] != 0:
        dat_dated.iloc[i,1:] /= maximums[i]
    else:
        pass
```

```python
fft = np.fft.rfft(dat_dated.iloc[:,1:])
```

```python
emb_umap = umap.UMAP(verbose=2).fit_transform(abs(fft))
```

```python
plt.figure()
plt.scatter(emb_umap[:,0],emb_umap[:,1],s=10,c = np.log(maximums),cmap='plasma')
```
